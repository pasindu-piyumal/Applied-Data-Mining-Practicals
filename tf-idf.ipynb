```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
d0 = "Department of Software Engineering in faculty of computing"
d1 = "Department of Data Science in faculty of computing"
d2 = "Department of Information Systems in faculty of computing"
String = [d0, d1, d2]
```

```python
import pandas as pd
```

```python
tfidf = TfidfVectorizer()
result = tfidf.fit_transform(String)
feature_names = tfidf.get_feature_names_out()
df = pd.DataFrame(result.toarray(), columns = feature_names)
print(df)
```
